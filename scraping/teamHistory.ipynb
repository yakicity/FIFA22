```python
import bs4
import re 
import requests
from urllib import request
import pandas as pd
import numpy as np
import time
```

```python
def teamHistory(lastYear, baseUrl):
    headers = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36'}
    allVals = []
    cols = []
    years = []
    for year in range(lastYear,1965,-1):
        print(year,end = ' ')
        url = baseUrl + str(year)
        response = requests.get(url=url, headers=headers)
        soup = bs4.BeautifulSoup(response.text,'html.parser')
        response.close()

        vals = []
        tableCols = soup.findAll('th', attrs={ 'class': 'zentriert' })
        tableVals = soup.findAll('td', attrs={ 'class': 'zentriert' })
        tableHeader = soup.findAll('div', attrs={ 'class': 'table-header' })
        if(tableHeader == []):
            continue
        if(year == lastYear):
        # cols.append('Year')
            cols.append('League')
            for j in range(1,7):
                cols.append('Home :' + str(tableCols[j].contents[0]))
            for j in range(1,7):
                cols.append('Away :' + str(tableCols[j].contents[0]))
            for j in range(1,7):
                cols.append('Overall :' + str(tableCols[j].contents[0]))
        vals.append(tableHeader[0].contents[1].contents[0]) 
    # たぶん一番上のテーブルの列数は6だから６＊３で変わらなさそう
        try:
            for i in range(0,18):
                vals.append(tableVals[i].contents[0])
        except IndexError:
            continue
        years.append(year)
        allVals.append(vals)
        time.sleep(5)
    return allVals,cols,years
```

```python
lastYear = 2020
baseUrl = 'https://www.transfermarkt.com/manchester-city/spielplan/verein/281/saison_id/'

result =  teamHistory(lastYear,baseUrl)
```

```python
df = pd.DataFrame(
    data=np.array(result[0]).T,
    index=result[1],
    columns=result[2]
)
df = df.T
# df.head()
df.to_csv('ManchesterCity_Year.csv')
```
