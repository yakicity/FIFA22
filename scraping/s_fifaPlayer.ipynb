```python
import bs4
import re 
import requests
from urllib import request
import pandas as pd
import numpy as np
import time
```

```python
# まとめ

headers = {'User-agent': 'Mozilla/5.0'}

statusdatas = []
playerNames = []
abilityNames = []
  
# 60人毎に選手が載ってるページを60回繰り返す
for i in range(0,20):
  # リンクの情報を取ってくる
  link = i * 60
  url = 'https://sofifa.com/players/?hl=ja-JP&col=oa&sort=desc&offset=' + f'{link}'
  time.sleep(5)
  response = requests.get(url=url, headers=headers)
  # 60人のページをsoupとする
  soup = bs4.BeautifulSoup(response.text,'html.parser')
  response.close()

  if(i == 0):
    # 初回のページからのみ能力値の名前をとってくる
    abilities = soup.findAll('label', attrs={ 'class': 'ellipsis' })
    abilityNames = [ability.contents[0] for ability in abilities[9:43]]

  # 各プレイヤー詳細ページのリンクをとってくる
  players = soup.findAll('tr')
  playerlink = []
  # for p in players:
  for j in range(1,61):
    p = players[j]
    tds = p.findAll('td')
    s = 'https://sofifa.com' + str(tds[1].contents[1].attrs['href'])
    playerlink.append(s)
  
  for j in range(0,60):
    time.sleep(5)
    # 各プレイヤー
    response2 = requests.get(url=playerlink[j], headers=headers)
    soup2 = bs4.BeautifulSoup(response2.text,'html.parser')
    response2.close()
    playerNames.append(str(soup2.findAll('h1')[1])[4:][:-5])
    status = soup2.findAll('span', attrs={ 'class': re.compile(r'bp3-tag p p-.*') })
    # あるプレイヤーの能力値を格納する配列
    statusdata = []
    for k in status:
      statusdata.append(float(k.contents[0]))  
    # 必要な能力値
    usefulStats = statusdata[-34:]

    # 他の情報も各プレイヤーごとにとってくる
    info1 = soup2.findAll('div', attrs={ 'class': 'block-quarter' })
    info2 = soup2.findAll('div', attrs={ 'class': 'meta ellipsis' })
    # 初回のページかつ初回プレイヤーからのみ情報の名前をとってくる
    if(i == 0 and j == 0):
      abiNamesAnother = ['Nationality', 'Birth', 'Height', 'Weight', 'Position', 'Overall Rating', 'Potential', 'Value', 'Wage', 'Preferred Foot', 'Weak Foot', 'Skill Moves']
      abilityNames = abilityNames + abiNamesAnother
    usefulStats.append(info2[0].contents[0].attrs['title'])
    # デバッグ用
    print(j, end='')
    length = len(info2[0])
    info2Split = str(info2[0].contents[length -1]).split(' ')
    usefulStats.append(info2Split[2][1:] + ',' + info2Split[3] + info2Split[4][:-1])
    usefulStats.append(info2Split[5])
    usefulStats.append(info2Split[6])
    usefulStats.append(info2[0].contents[2].contents[0])
    usefulStats.append(info1[0].contents[0].contents[0].contents[0])
    usefulStats.append(info1[1].contents[0].contents[0].contents[0])
    usefulStats.append(info1[2].contents[0].contents[0])
    usefulStats.append(info1[3].contents[0].contents[0])
    usefulStats.append(info1[4].contents[1].contents[3].contents[1].contents[1])
    usefulStats.append(info1[4].contents[1].contents[3].contents[3].contents[0])
    usefulStats.append(info1[4].contents[1].contents[3].contents[5].contents[0])
    statusdatas.append(usefulStats)
```

```python
df = pd.DataFrame(
    data=np.array(statusdatas).T,
    index=abilityNames,
    columns=playerNames
)
df = df.T
df.head()
```

```python
df.to_csv('1200.csv')
```
