```python
import csv
import bs4
import re 
import requests
from urllib import request
import pandas as pd
import numpy as np
import time
```

```python
headers = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36'}
```

```python
def teamPerMatch(team, year, baseUrl):
    url = baseUrl + str(year)
    response = requests.get(url=url, headers=headers)
    soup = bs4.BeautifulSoup(response.text,'html.parser')
    response.close()

    matchesInfo = []
    cols = []

    getlink = soup.findAll('a', attrs={ 'class': 'ergebnis-link' })
    links = []
    # シーズンすべての試合のリンク
    for i in range(len(getlink)):
      links.append('https://www.transfermarkt.com' + str(getlink[i].attrs['href']))

    dateHeld = soup.findAll('td', attrs={ 'class': 'zentriert' })
    matchesInfo = []
    for match in range(len(links)):
      matchInfo = []
      # manCityのホーム戦かどうか
      isHome = False
      if(len(dateHeld[18 + match*8]) == 1):
        matchInfo.append(str(dateHeld[18 + match*8].contents[0]).strip(' \n '))
      if(len(dateHeld[18 + match*8]) == 3):
        matchInfo.append(str(dateHeld[18 + match*8].contents[1].contents[0]).strip(' \n '))
      matchInfo.append(str(dateHeld[19 + match*8].contents[0]).strip(' \n ').split(',')[0])
      matchInfo.append(str(dateHeld[20 + match*8].contents[0]).strip(' \n '))
      matchInfo.append(dateHeld[21 + match*8].contents[0])
      if( dateHeld[21 + match*8].contents[0] == 'H'):
        isHome = True
      # rankingはカップ戦は存在しない
      try:
        matchInfo.append(re.search('[0-9]',str(dateHeld[22 + match*8].contents[0].contents[0])).group(0))
      except:
        matchInfo.append('-')
      matchInfo.append(dateHeld[23 + match*8].contents[0].contents[0].get('alt'))
      matchInfo.append(dateHeld[24 + match*8].contents[0])
      matchInfo.append(dateHeld[25 + match*8].contents[0].contents[0].contents[0])
      matchesInfo.append(matchInfo)
      print(match, end= ' ')

      # それぞれの試合ごとの情報をとってくる
      link = links[match]

      # indexページ
      responsePerMatch = requests.get(url=link, headers=headers)
      time.sleep(3)
      soupPerMatch = bs4.BeautifulSoup(responsePerMatch.text,'html.parser')
      responsePerMatch.close()

      # 前半での得点
      getHalfScore = soupPerMatch.findAll('div', attrs={ 'class': 'sb-halbzeit' })
      halfScore = str(getHalfScore[0].contents[1].contents[0]) + str(getHalfScore[0].contents[2])[:-1]
      matchesInfo[match].append(halfScore)

      # statisticsページとclubComparisonページのリンクを取得
      otherLink = soupPerMatch.findAll('a', attrs={ 'class': 'tm-subnav-item megamenu' })
      # statisticsページとclubComparisonページがない場合はそのデータはすべて' - '
      tabs = []
      for i in range(len(otherLink)):
        tabs.append(otherLink[i].contents[0])

      # statisticsページ
      try:
        statisticsLink = 'https://www.transfermarkt.com' + otherLink[tabs.index('Statistics')].attrs['href']
        responseStatistics = requests.get(url=statisticsLink, headers=headers)
        time.sleep(3)
        soupStatistics = bs4.BeautifulSoup(responseStatistics.text,'html.parser')
        responseStatistics.close()
        # 審判Referee
        matchesInfo[match].append(soupStatistics.findAll('p',attrs={ 'class': 'sb-zusatzinfos' })[0].contents[6].contents[0])
        # statisticを追加
        if(isHome):
          for i in range(7):
            matchesInfo[match].append(soupStatistics.findAll('li', attrs={ 'class': 'sb-statistik-heim' })[i].contents[1].contents[3].contents[0])
        else:
          for i in range(7):
            matchesInfo[match].append(soupStatistics.findAll('li', attrs={ 'class': 'sb-statistik-gast' })[i].contents[1].contents[3].contents[0])
        if(~isHome):
          for i in range(7):
            matchesInfo[match].append(soupStatistics.findAll('li', attrs={ 'class': 'sb-statistik-heim' })[i].contents[1].contents[3].contents[0])
        else:
          for i in range(7):
            matchesInfo[match].append(soupStatistics.findAll('li', attrs={ 'class': 'sb-statistik-gast' })[i].contents[1].contents[3].contents[0])
      except:
        for i in range(14):
          matchesInfo[match].append(' - ')

      # clubComparisonページ
      try:
        clubCompsLink = 'https://www.transfermarkt.com' + otherLink[tabs.index('Club Comparison')].attrs['href']
        time.sleep(3)
        responseClubComps = requests.get(url=clubCompsLink, headers=headers)
        soupClubComps = bs4.BeautifulSoup(responseClubComps.text,'html.parser')
        responseClubComps.close()
        # clubComparisonを追加
        if(isHome):
          for i in range(7):
            matchesInfo[match].append(soupClubComps.findAll('span', attrs={ 'class': 'datenundfakten_bar' })[2 * i + 1].contents[0])
        else:
          for i in range(7):
            matchesInfo[match].append(soupClubComps.findAll('span', attrs={ 'class': 'datenundfakten_bar' })[2 * i].contents[0])
        if(~isHome):
          for i in range(7):
            matchesInfo[match].append(soupClubComps.findAll('span', attrs={ 'class': 'datenundfakten_bar' })[2 * i].contents[0])
        else:
          for i in range(7):
            matchesInfo[match].append(soupClubComps.findAll('span', attrs={ 'class': 'datenundfakten_bar' })[2 * i + 1].contents[0])
      except:
        for i in range(14):
          matchesInfo[match].append(' - ')


      # 得点数のカウント用
      getGoalPlayers = soupPerMatch.findAll('div', attrs={ 'class': 'sb-ereignisse' })
      # 得点、交代、カードの選手一覧
      getInfo = soupPerMatch.findAll('div', attrs={ 'class': 'sb-aktion-aktion' })
      # ヘッダー
      getHead = soupPerMatch.findAll('div', attrs={ 'class': 'header' })
      header = []
      for i in range(2,len(getHead)):
        header.append(getHead[i].contents[1].contents[0])
      eachNum = []
      for i in range(len(getGoalPlayers)):
        eachNum.append(len(getGoalPlayers[i].contents[1])//2)
      # ヘッダーとその数の辞書を作る
      headerDict = {}
      for i in range(len(eachNum)):
        headerDict[header[i]] = eachNum[i]

      try:
        goals = headerDict['Goals']
        goalI= header.index('Goals')
      except:
        goals = 0

      try:
        penalties = headerDict['missed penalties']
        penI= header.index('missed penalties')
      except:
        penalties = 0

      try:
        subs = headerDict['Substitutions']
        subI= header.index('Substitutions')
      except:
        subs = 0

      try:
        cards = headerDict['Cards']
        cardI= header.index('Cards')
      except:
        cards = 0


      goalInfo = ''
      if(goals != 0):
        for i in range(goals):
          if(getGoalPlayers[goalI].contents[1].contents[2*i+1].attrs['class'][0] == 'sb-aktion-heim'):#home
            if(isHome):
              goalInfo = goalInfo + team + ' | '
            else:
              goalInfo = goalInfo + 'enemy' + ' | '
          else:
            if(isHome):
              goalInfo = goalInfo + 'enemy' + ' | '
            else:
              goalInfo = goalInfo + team + ' | ' 
          x = 1 + int(re.findall(r'\d+', getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[1].contents[0].attrs['style'])[0]) //36#=>9
          y = int(re.findall(r'\d+',getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[1].contents[0].attrs['style'])[1]) // 36 #=>4
          minute = y * 10 + x#=>49
          goalInfo += str(minute) + ' | '
          goalInfo += str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[1].contents[0]) + ' | ' #=>Ilkay Gündogan
          # goalInfo += str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[2]) + ' | ' #=>, Right-footed shot, 8. Goal of the Season
          try:
            goalReason = str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[2]).split(',')[1].strip( )
            goalInfo += goalReason + ' | ' #=>Right-footed shot
          except:
            goalInfo += '-' + ' | ' #=>-      
          try:
            goalInfo += str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[2]).split(',')[2].strip( ) + ' | ' #=>8. Goal of the Season
          except:
            goalInfo += '-' + ' | '
          # アシストある例
          if(goalReason == 'Penalty'):
            goalInfo += 'No Assist | '
          else:
            try:
              goalInfo += str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[5].contents[0]) + ' | ' #=>Phil Foden
              # goalInfo += str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[6]) + ' | ' #=>, Pass, 3. Assist of the Season
              try:
                goalInfo += str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[6]).split(',')[1].strip( ) + ' | ' #=>Pass
              except:
                goalInfo += '-' + ' | ' #=>-
              try:
                goalInfo += str(getGoalPlayers[goalI].contents[1].contents[2*i+1].contents[1].contents[7].contents[6]).split(',')[2].strip( ) + ' | ' #=>3. Assist of the Season  
              except:
                goalInfo += '-' + ' | ' #=>-
            except:
              goalInfo += 'No Assist | '
          goalInfo += ' @'
      else:
        goalInfo += 'No Goal @'

      subInfo = ''
      if(subs != 0):
        for i in range(subs):
          if(getGoalPlayers[subI].contents[1].contents[2*i+1].attrs['class'][0] == 'sb-aktion-heim'):#home
            subInfo = subInfo + 'home' + ' | '
          else:
            subInfo += 'away' + ' | '
          x = 1 + int(re.findall(r'\d+', getGoalPlayers[subI].contents[1].contents[2*i+1].contents[1].contents[1].contents[0].attrs['style'])[0]) //36#=>9
          y = int(re.findall(r'\d+',getGoalPlayers[subI].contents[1].contents[2*i+1].contents[1].contents[1].contents[0].attrs['style'])[1]) // 36 #=>4
          minute = y * 10 + x#=>49
          subInfo += str(minute) + ' | '
          subInfo += getGoalPlayers[subI].contents[1].contents[2*i+1].contents[1].contents[7].contents[5].contents[1].contents[0] + ' | '#=>Riyad Mahrezが下がる
          subInfo += getGoalPlayers[subI].contents[1].contents[2*i+1].contents[1].contents[7].contents[3].contents[3].contents[0] + ' | '#=>Gabriel Jesusがあがる
          subReason = str(getGoalPlayers[subI].contents[1].contents[2*i+1].contents[1].contents[7].contents[5].contents[2].contents[0])[2:-1]
          if(subReason != ''):
            subInfo += subReason + ' | '#=>Injury
          else:
            subInfo += '-' + ' | '
          subInfo += ' @' 
      else:
        subInfo += 'No Sub @' 

      cardInfo = ''
      if(cards != 0):
        for i in range(cards):
          if(getGoalPlayers[cardI].contents[1].contents[2*i+1].attrs['class'][0] == 'sb-aktion-heim'):#home
            cardInfo = cardInfo + 'home' + ' | '
          else:
            cardInfo += 'away' + ' | '
          x = 1 + int(re.findall(r'\d+', getGoalPlayers[cardI].contents[1].contents[2*i+1].contents[1].contents[1].contents[1].attrs['style'])[0]) //36#=>9
          y = int(re.findall(r'\d+',getGoalPlayers[cardI].contents[1].contents[2*i+1].contents[1].contents[1].contents[1].attrs['style'])[1]) // 36 #=>4
          minute = y * 10 + x#=>49
          cardInfo += str(minute) + ' | '
          cardInfo += getGoalPlayers[cardI].contents[1].contents[2*i+1].contents[1].contents[7].contents[0].contents[0]+ ' | '#=>Thiago
          try:  
            cardInfo += str(getGoalPlayers[cardI].contents[1].contents[2*i+1].contents[1].contents[7].contents[2]).split(',')[0].strip( )+ ' | ' #=>1. Yellow card
          except:
            cardInfo += '-' + ' | ' #=>-
          try:
            cardInfo += str(getGoalPlayers[cardI].contents[1].contents[2*i+1].contents[1].contents[7].contents[2]).split(',')[1].strip( )+ ' | '  #=>Foul
          except:
            cardInfo += '-' + ' | ' #=>-
          cardInfo += ' @'
      else:
        cardInfo += 'No Card @' 

      matchesInfo[match].append(goalInfo)
      matchesInfo[match].append(subInfo)
      matchesInfo[match].append(cardInfo)
    
    cols = ['Matchday','Date','Time','Venue','Ranking','Opponent','System','Result','HalfScore', 'Referee']
    for i in range(1,8):
      cols.append(str(soupStatistics.findAll('div', attrs={ 'class': 'unterueberschrift' })[i].contents[0])[4:].strip(' '))
    for i in range(1,8):
      cols.append('ENEMY ' + str(soupStatistics.findAll('div', attrs={ 'class': 'unterueberschrift' })[i].contents[0])[4:].strip(' '))
    cols.append('Total market value')
    for i in range(1,7):
      cols.append(soupClubComps.findAll('td', attrs={ 'class': 'daten-und-fakten-kategorie' })[i].contents[1].contents[0])
    cols.append('ENEMY Total market value')
    for i in range(1,7):
      cols.append('ENEMY ' + str(soupClubComps.findAll('td', attrs={ 'class': 'daten-und-fakten-kategorie' })[i].contents[1].contents[0]))
    cols.append('GoalInfo')
    cols.append('SubInfo')
    cols.append('CardInfo')
    
    return matchesInfo,cols,len(links)
```

**特殊なやつ…なんかの要素が抜けてそうなやつ**\
 'away | 55 | Riyad Mahrez | Penalty | 1. Tournament Goal | Emre Can |  ',\
  'home | 90 | Álvaro González | Own-goal | - |  ', \
**基本的にはPK戦とかいれたかったら、「ヘッダーとその数の辞書」からそれを探して、他のと同様にPK専用のコードを書いてif文でそこに飛ばせばたぶんいけるはず。今のところはゴール、交代、か―ドだけデータ取ってきてる。**

```python
team = 'ManC'
year = 2020
baseUrl = 'https://www.transfermarkt.com/manchester-city/spielplan/verein/281/saison_id/'
result = teamPerMatch(team, year, baseUrl)
```

```python
df = pd.DataFrame(
    data=result[0],
    index=range(result[2]),
    columns=result[1]
)
df
```

```python
df.to_csv('teamPerMatch.csv')
```
